```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.firefox.service import Service
import time
import pandas as pd
import random 
import csv 
import math
from pathlib import Path
```

```python
option = webdriver.FirefoxOptions()
option.set_preference("browser.download.manager.showWhenStarting", False)
option.set_preference('browser.helperApps.neverAsk.saveToDisk',
                       'application/vnd.openxmlformats-officedocument.spreadsheetml.sheet'
                       ';application/vnd.ms-excel'
                       ';application/xls'
                       ';application/xlsx')
option.set_preference("permissions.default.stylesheet", 2)
#option.set_preference("permissions.default.image", 2)
option.set_preference("javascript.enabled", True)
```

```python
district_url = ['https://dmft.karnataka.gov.in/Bidar/', 'https://dmft.karnataka.gov.in/Kalburgi/', 'https://dmft.karnataka.gov.in/Vijayapura/', 'https://dmft.karnataka.gov.in/Yadgir/', 'https://dmft.karnataka.gov.in/Raichur/', 'https://dmft.karnataka.gov.in/Bagalkot/','https://dmft.karnataka.gov.in/Belagavi/', 'https://dmft.karnataka.gov.in/Dharwad/', 'https://dmft.karnataka.gov.in/Gadag/', 'https://dmft.karnataka.gov.in/Koppal/', 'https://dmft.karnataka.gov.in/Ballari/', 'https://dmft.karnataka.gov.in/Vijayanagara/', 'https://dmft.karnataka.gov.in/Haveri/', 'https://dmft.karnataka.gov.in/UttaraKannada/', 'https://dmft.karnataka.gov.in/Shivamogga/', 'https://dmft.karnataka.gov.in/Davanagere/', 'https://dmft.karnataka.gov.in/Udupi/', 'https://dmft.karnataka.gov.in/Chikkamagaluru/', 'https://dmft.karnataka.gov.in/Chitradurga/', 'https://dmft.karnataka.gov.in/DakshinaKannada/','https://dmft.karnataka.gov.in/Hassan/', 'https://dmft.karnataka.gov.in/Tumakuru/', 'https://dmft.karnataka.gov.in/Chikkaballapura/', 'https://dmft.karnataka.gov.in/BengaluruRural/', 'https://dmft.karnataka.gov.in/BengaluruUrban/', 'https://dmft.karnataka.gov.in/Ramanagar/', 'https://dmft.karnataka.gov.in/Mandya/', 'https://dmft.karnataka.gov.in/Mysuru/', 'https://dmft.karnataka.gov.in/Kodagu/', 'https://dmft.karnataka.gov.in/Kolara/', 'https://dmft.karnataka.gov.in/Chamarajanagara/' ]
```

```python
master_df = pd.DataFrame(index=[], columns= [])

district_name =[]

import re #for pattern matching import regex 

for i in district_url: 
    #get the district name from url 
    sub_temp_name = i[30:]
    temp_name = re.sub("/", "", sub_temp_name)
    #print(temp_name
    district_name.append(temp_name)

    temp_dist_url = i+"/"+ "dmftcollection" 


    #now open an url
    driver= webdriver.Firefox(options=option)
    driver.maximize_window() 
    driver.get(temp_dist_url)
    driver.implicitly_wait(10)

    #element1 = driver.find_element(By.TAG_NAME, "body")

    #scroll_height = element1.get_attribute("scrollHeight")
    #print(scroll_height)
    driver.execute_script("window.scrollTo(130, 170);")


    for k in range(2, 9,1): 
        #no loop execution 
        a = ActionChains(driver)        
        element = driver.find_element(By.XPATH, '/html/body/app-root/div[2]/app-dmft/section[2]/div[2]/div/div/table/tbody/tr['+str(k)+']/td[6]/a')
        a.move_to_element(element)
        a.click(on_element= element).perform()
        time.sleep(8)
       #get the table data 
        tables = WebDriverWait(driver, 20).until(ec.presence_of_all_elements_located((By.XPATH, '//*[@id="monthwise_modal"]/div/div/div/div/table')))

        for table in tables:
            df_st= pd.read_html(table.get_attribute('outerHTML'))

        temp_df= df_st[0]
    
        len_temp = len(temp_df.index)

        temp_d_name = [temp_name]*len_temp

        temp_df['District'] = temp_d_name  

        print(temp_df.head(5))

        master_df = pd.concat([master_df, temp_df])

        a.reset_actions()

        close_button = driver.find_element(By.CLASS_NAME, "btn-close")

        a.click(on_element = close_button)

        time.sleep(2)
    driver.close()
    time.sleep(2)
```
