```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.firefox.service import Service
import time
import pandas as pd
import random 
import csv 
import math
from pathlib import Path
```

```python
option = webdriver.FirefoxOptions()
option.set_preference("browser.download.manager.showWhenStarting", False)
option.set_preference('browser.helperApps.neverAsk.saveToDisk',
                       'application/vnd.openxmlformats-officedocument.spreadsheetml.sheet'
                       ';application/vnd.ms-excel'
                       ';application/xls'
                       ';application/xlsx')
option.set_preference("permissions.default.stylesheet", 2)
option.set_preference("permissions.default.image", 2)
option.set_preference("javascript.enabled", True)
```

```python
driver= webdriver.Firefox(options=option)
driver.maximize_window()
url = "https://dmf.gujarat.gov.in/Home/BoardOfTrusteeList?memberType=v8w6LzK9ededwzFBZm4rOQ%3D%3D"
driver.get(url)

assert "DMF" in driver.title
time.sleep(1)




```

```python
committee_path = driver.find_elements(By.XPATH, '//*[@id="membertype"]/option[position()>1]')

committee_name = [m.get_attribute('textContent') for m in committee_path]

master_df = pd.DataFrame( columns=['SL', 'District', 'Committee', 'Name', 'Designation of Member in Committee'])

for j in range(0, len(committee_name), 1): 
    driver.find_element(By.XPATH, '//*[@id="membertype"]/option['+str(j+2)+']').click()
    district_path = driver.find_elements(By.XPATH, '//*[@id="district"]/option[position()>2]')
    district_name = [m.get_attribute("textContent") for m in district_path]
    for i in range(0, len(district_name), 1):
        driver.find_element(By.XPATH, '//*[@id="district"]/option[' +str(i+2)+']').click()
        ##click on show button 
        driver.find_element(By.XPATH, '//*[@id="btnshow"]').click()

        tables = WebDriverWait(driver, 20).until(ec.presence_of_all_elements_located((By.XPATH, '//*[@id="example"]')))
        for table in tables:
           lst= pd.read_html(table.get_attribute('outerHTML'))

        temp_df= lst[0]

        #print(temp_df.head(5))

        master_df = pd.concat([master_df, temp_df])
        print(master_df.tail(5))

        path = "C:/Users/niranjan.kumar/Desktop/DMFT_Scraper/"

        master_df.to_csv(path+'DMFT_Member_Gujarat.csv', index=False)



read_csv_saved = pd.read_csv("C:/Users/niranjan.kumar/Desktop/DMFT_Scraper/DMFT_Member_Gujarat.csv")
        
```
