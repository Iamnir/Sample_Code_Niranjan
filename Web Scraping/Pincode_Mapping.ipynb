```python
import os
import pandas as pd

# Step 1: Define the folder path
folder_path = r'C:\\Users\\niranjan.kumar\\Desktop\\AppraisalGap\\Delhi_Dispute'

# Step 2: List all Excel files in the folder
excel_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]

# Step 3: Read each Excel file into a DataFrame and store them in a list
dfs = []
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    df = df.drop(columns=["SRO_NAME"])
    dfs.append(df)

# Step 4: Concatenate all DataFrames into a single DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)

column_mapping = {
    'Sr. No.': 'Sr__No_',
    'Complaint Type': 'Complaint_Type',
    'SRO Name': 'SRO_Name',
    'Corp Name': 'Corp_Name',
    'Property Address': 'Property_Address',
    'Date of Dispute': 'Date_of_Dispute'
}


concatenated_df = concatenated_df.rename(columns=column_mapping)

# Split the 'SRO_NAME' column based on the '-' character and expand it into separate columns
split_names = concatenated_df['SRO_Name'].str.split('-', expand=True)

# Assign the first part (before '-') to 'SRO_NAME' column and the second part (after '-') to 'SRO_AREA' column
concatenated_df['SRO_Category'] = split_names[0]
concatenated_df['SRO_AREA'] = split_names[1]

# Iterate over each column in the DataFrame
for column in concatenated_df.columns:
    # Convert the column to string data type
    concatenated_df[column] = concatenated_df[column].astype(str)

# Step 5: Save the concatenated DataFrame as a .dta file
output_file_path = os.path.join(folder_path, 'final_raw_data.dta')
concatenated_df.to_stata(output_file_path, version=118)
```

```python
from geopy.geocoders import Nominatim
import reverse_geocoder as rg
import swifter 

def locality_to_pincode(locality, sro_area, district):
    try:
        # Concatenate the address components
        address = f"{locality}, {sro_area}, {district}, Delhi, India"
        
        # Geocode the address to get coordinates
        geolocator = Nominatim(user_agent="geoapiExercises")
        location = geolocator.geocode(address)
        
        if location:
            # Use reverse geocoding to get pincode from coordinates
            lat, long = (location.latitude, location.longitude)
            location = geolocator.reverse((lat, long))

            #print(location)
            address_elements = location.address.split(", ")
            zipcode = address_elements[-2]
            print(zipcode)
            return zipcode 

    except Exception as e:
        print(f"An error occurred: {e}")
    
    return None

# Apply the locality_to_pincode function to concatenated_df to create a new column 'Pincode'

# Apply the locality_to_pincode function to concatenated_df to create a new column 'Pincode'
concatenated_df['Pincode'] = concatenated_df.swifter.apply(lambda row: locality_to_pincode(row['Locality'], row['SRO_AREA'], row['District']), axis=1)

```
